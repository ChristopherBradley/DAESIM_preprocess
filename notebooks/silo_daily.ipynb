```python
from DAESIM_preprocess.silo_daily import silo_daily , silo_abbreviations
```

# Basic Example

```python
# Example variables
variables = ['min_temp', 'max_temp']

lat = -34.3890427
lon=148.469499
buffer = 0.1  # Degrees in each direction (so ~20kmx20km)

start_year = 2020
end_year = 2021
```

```python
# The main function
ds = silo_daily(variables, lat, lon, buffer, start_year, end_year)
```

```python
# Example xarray that gets returned
ds
```

# More examples

```python
# Retrieving a single variable
print(f"Options: {list(silo_abbreviations.keys())}\n")
ds = silo_daily(variables=['monthly_rain'], lat=lat, lon=lon, buffer=buffer, start_year=start_year, end_year=end_year)
ds
```

```python
# Retrieving a single point
ds = silo_daily(variables=variables, lat=lat, lon=lon, buffer=0, start_year=start_year, end_year=end_year)
ds
```

```python
# Retrieving a single year
ds = silo_daily(variables=variables, lat=lat, lon=lon, buffer=buffer, start_year=2020, end_year=2020)
ds
```

```python
# Giving the files a new name
ds = silo_daily(variables=variables, lat=lat, lon=lon, buffer=buffer, start_year=start_year, end_year=end_year,
                 stub="NEW_STUB")
```

```python
# Saving the results in a new location
import os
if not os.path.exists('tmpdir'):
    os.mkdir('tmpdir')
if not os.path.exists('outdir'):
    os.mkdir('outdir')
ds = silo_daily(variables=variables, lat=lat, lon=lon, buffer=buffer, start_year=start_year, end_year=end_year, outdir='outdir', tmpdir='tmpdir')
```

```python
# Skipping the plot
ds = silo_daily(variables=variables, lat=lat, lon=lon, buffer=buffer, start_year=start_year, end_year=end_year, 
                  plot=False)
ds
```

```python
# Skipping the write to file of the final .nc 
# This is for performance to avoid write/reading if you use this object in the next step
# Note that the function still has to download a few smaller files because these get aggregated into the final xarray
ds = silo_daily(variables=variables, lat=lat, lon=lon, buffer=buffer, start_year=start_year, end_year=end_year, 
                  save_netcdf=False, plot=False)
ds
```
